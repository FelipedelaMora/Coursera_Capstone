Now I am importing wikipedia and Pandas library to parse Wikipedia webpage easily, this could also be done using BeutifulSoup, but since we are scraping a wikipedia page, the Wikipedia library males it really easy to extract information. Here is the [link](https://pypi.org/project/wikipedia/) to Python's official documentation on this library. 
You'll see that this will be parsed and converted into a pandas dataframe, which will be easier to manipulate later on.

```python
import pandas as pd
import wikipedia as wp
 
#Get the html source
html = wp.page("List_of_postal_codes_of_Canada:_M").html().encode("UTF-8")
df = pd.read_html(html)[0]
df.head()
```

On this step, I will be finding and removing all rows that contain a **'Not assigned'** Borough, keeping the inplace value as True, to make the changes on the dataframe.

```python
#Removing 'Not Assigned' boroughs
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
df.rename(columns={'Postcode':'Postal Code'}, inplace=True)
df
```

I will apply a simple groupby into the dataframe, grouping it by **Postcode** and **Borough**, excluding the **Neighbourhood** so I can apply a join method to put together values on the same row, separated by a comma.

```python
#Merging duplicate rows and joining values
df2 = df.groupby(['Postal Code','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df2
```

After the grouping, I will search for any Neighbourhood that containts **Not asssigned** and make it equal to the equivalent **Borough** value. 

```python
df2.loc[df2.Neighbourhood == 'Not assigned', 'Neighbourhood'] = df2.Borough.shift(0)
df2
```

Now let's see the number of columns and rows on our dataframe.

```python
df2.shape
```

Now I am defining the number of venues to retrieve and the radius to search for in to simple variable, which I will use later on.

On this cell we are defining a function to get 100 venues for each **Neighbourhood** on a 500 radius. 

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Let's execute the function we created.

We will need to encode out dataframe in order to fit it into an statistical algorithm, easiest way to do this is pandas **get_dummies** method. Let's apply it. 

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

Now we have a dataframe that is encoded, with a column for each venue category, so let's group it to make it easier to understand and apply the mean method to identify frequencies.

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

We are going to analyze only the **top 5** venues, therefore, we are going to extract them for each neighbourhood with a for loop. 

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

By defining a function to extract the most commong venues in a descending way, we will give them a dataframe format. 

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Once the function is created, we move into the dataframe creation. 

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

On this section, we will start the K-Mean algorithm to classify and cluster venues where we are going to fit our dataframe. On this case, we are defining only 5 clusters to be created. You can play with it if you want to see changes. 

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Now that we have clusters created, let's assign them labels so we can identify them easily. 

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

toronto_merged.head() # check the last columns!
```

Curious on how our clusters look like on a map? Let's find out!

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
